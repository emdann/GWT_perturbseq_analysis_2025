```python
import pandas as pd
import numpy as np
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
import os,sys
import anndata as ad

sys.path.append(os.path.abspath('../'))
import utils
```

```python
base_data_dir = '../../../../3_expts/Diff043scRNAseq'

all_anndatas = []

timepoint = ['Rest', 'Stim8hr', 'Stim48hr']
donor = ['D1', 'D2', 'D3', 'D4']
condition = ['Mix', 'Th0', 'Th1', 'Th2']

for tp in timepoint:
    for dn in donor:
        for ct in condition:
            h5_file_path = base_data_dir + '/Diff043' + tp + '/per_sample_outs/' + tp + '_' + dn + '_' + ct + '/count/sample_filtered_feature_bc_matrix.h5'
            print(f"Loading {h5_file_path}...")
            adata = sc.read_10x_h5(h5_file_path)
            adata.obs['timepoint'] = tp
            adata.obs['donor'] = dn
            adata.obs['culture_condition'] = ct
            all_anndatas.append(adata)
```

```python
for adata in all_anndatas:
    base_original_index = adata.obs.index.str[:-2]
    new_index = (
        base_original_index + '_' +
        adata.obs['timepoint'].astype(str) + '_' +
        adata.obs['donor'].astype(str) + '_' +
        adata.obs['culture_condition'].astype(str)
    )
    adata.obs.index = new_index

combined_anndata = ad.concat(all_anndatas, join='outer')
combined_anndata.var = adata.var[['gene_ids']]
combined_anndata.var['gene_name'] = adata.var.index
```

```python
sc.pp.filter_cells(combined_anndata, min_genes=200)
```

```python
combined_anndata.write_h5ad(base_data_dir+'/Diff043_merged.gex.filtered.h5ad')
```

```python
# Separate data based on culture conditions and save separately
adata_Mix = combined_anndata[combined_anndata.obs.culture_condition=='Mix'].copy()
adata_Mix.write_h5ad(base_data_dir+'/Diff043_Mix_merged.gex.filtered.h5ad')
adata_Th0 = combined_anndata[combined_anndata.obs.culture_condition=='Th0'].copy()
adata_Th0.write_h5ad(base_data_dir+'/Diff043_Th0_merged.gex.filtered.h5ad')
adata_Th1 = combined_anndata[combined_anndata.obs.culture_condition=='Th1'].copy()
adata_Th1.write_h5ad(base_data_dir+'/Diff043_Th1_merged.gex.filtered.h5ad')
adata_Th2 = combined_anndata[combined_anndata.obs.culture_condition=='Th2'].copy()
adata_Th2.write_h5ad(base_data_dir+'/Diff043_Th2_merged.gex.filtered.h5ad')
```
